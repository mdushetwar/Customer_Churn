```python
Q1 = d.quantile(0.25)
Q3 = d.quantile(0.75)
IQR = Q3 - Q1

outliers = ((d < (Q1 - 1.5 * IQR)) | (d > (Q3 + 1.5 * IQR))).sum() / len(d) * 100

print(outliers)
```

```python
categorical_col = []#list
for column in d.columns:#for loop to acess columns form dataset
    if d[column].dtype == object and len(d[column].unique()) <= 50:#checking datatype whether datatype is object/string and number of unique label in the columns less than 50 
        categorical_col.append(column)#appending those columns in the list who statisfy the condition 
        print(f"{column} : {d[column].unique()}")#output
        print("====================================")
```
